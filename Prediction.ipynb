```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df['Age'] = df['Age'].fillna(23.8)
```

```python
df.tail()
```

```python
df['Sex'] = df['Sex'].replace({'male': 0 ,'female':1})
```

```python
df.head()
```

```python
df.drop(columns=['Embarked','Cabin','Fare','Ticket','PassengerId'],inplace=True)
```

```python
df.drop(columns=['Name'],inplace=True)
```

```python
df.head()
```

```python
correlation = df.corr()
correlation['Survived'].sort_values(ascending=False)
```

```python
X = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
algo = DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=5,min_samples_leaf=5)
algo.fit(X_train,y_train)
pred = algo.predict(X_test)
print(round(accuracy_score(y_test,pred)*100,2))
pd.DataFrame({'Actual Survival':y_test,'Predicted Survival': pred})
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
algo2 = LinearRegression()
algo2.fit(X_train,y_train)
pred2 = algo.predict(X_test)
print(round(accuracy_score(y_test,pred2)*100,2))
pd.DataFrame({'Actual Survival':y_test,'Predicted Survival': pred2})
```
